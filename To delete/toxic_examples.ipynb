```python
import pandas as pd
import numpy as np
from detoxify import Detoxify

classifier = Detoxify('unbiased')
Examples = ['The monkey makes jokes', 'The mean monkey makes means jokes about means', 'The annoying monkey makes annoying jokes about means',
'The vile monkey makes vile jokes about means', 'The malevolent monkey makes malevolent jokes about means', 
'The nasty monkey makes nasty jokes about means', 'The despicable monkey makes despicable jokes about means', 
'The bastard monkey makes bastard jokes about means']
results = pd.DataFrame(classifier.predict(Examples))
results['Comments'] = Examples
results = results[['Comments', 'toxicity',  'severe_toxicity', 'obscene', 'identity_attack', 'insult', 'threat', 'sexual_explicit']]
results
```

```python
Examples_2 = ['I am at the zoo and I see a black monkey', 'I saw a black monkey at the zoo']
results_2 = pd.DataFrame(classifier.predict(Examples_2))
results_2['Comments'] = Examples_2
results_2 = results_2[['Comments', 'toxicity',  'severe_toxicity', 'obscene', 'identity_attack', 'insult', 'threat', 'sexual_explicit']]
results_2
```

**Comments**: the formulation of a sentence can make the toxicity vary a lot
